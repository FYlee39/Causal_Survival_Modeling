```python
import pandas as pd
import numpy as np

from helper_functions import *
from sklearn.model_selection import ParameterGrid

RANDOM_SEED = 42
```

```python

Global_param_grid = {"Time from admission": [True, False],
                     "Include treatment": [True, False],
                     "Condition on has_Sepsis_A41_9": [True, False],
                     "Use weighted": [True, False],
             }

Global_params = ParameterGrid(Global_param_grid)
```

```python
categorical_features = ["gender",
                        "ethnicity",
                        "admission_type",
                        "has_COPD",
                        "has_Diabetes",
                        "has_Metastasis",
                        "has_Sepsis_A41_9"]

DCM_param_grid = {"k" : [3],
                  "learning_rate" : [1e-3],
                  "layers": [[50, 50]],
                  "iters": [100]
             }
DCM_params = ParameterGrid(DCM_param_grid)
```

```python
np.random.seed(RANDOM_SEED)

model_list = []

index = 0

for param in Global_params:

    time = "admission" if param["Time from admission"] else "0"
    use_treatment = "include" if param["Include treatment"] else "exclude"
    condition = "data only has Sepsis_A41_9" if param["Condition on has_Sepsis_A41_9"] else "all data"
    weighted = "use weighted" if param["Use weighted"] else "no weighted"

    save_name = f"from {time}, {use_treatment} treatment, using {condition}, {weighted}_{index}"
    index += 1

    print(f"For setting: {save_name}")

    cleaned_df = processing_data(path="C:\\Users\\lee39\\OneDrive\\Desktop\\new_final_merged_dataset.csv",
                             time_from_admission=param["Time from admission"])

    if param["Condition on has_Sepsis_A41_9"]:
        cleaned_df = cleaned_df[cleaned_df["has_Sepsis_A41_9"] == 1]

    cleaned_df = cleaned_df[cleaned_df["time"] > 0]
    treatment = cleaned_df["treatment"]

    if param["Include treatment"]:
        processed_list = (categorical_features + ["treatment"])
    else:
        processed_list = categorical_features
        cleaned_df.drop("treatment", axis=1, inplace=True)

    X_data, t_data, e_data, categorical_features_list, numerical_features_list = (
        processing_data_2_DCM(df=cleaned_df,
                              categorical_features_list=processed_list,
                              train_test_val_size=(0.7, 0.2, 0.1),
                              random_seed=RANDOM_SEED,
                              clustering=False,
                              n_clusters=3,
                              )
    )

    X_train, X_val, X_test = X_data
    t_train, t_val, t_test = t_data
    e_train, e_val, e_test = e_data

    whole_X_df = pd.concat([X_train, X_val, X_test], axis=0)
    whole_t_df = pd.concat([t_train, t_val, t_test], axis=0)
    whole_e_df = pd.concat([e_train, e_val, e_test], axis=0)

    # get the name of covariates
    covariates = list(whole_X_df.columns)

    if param["Include treatment"]:
        whole_df_ps = whole_X_df
        treatment_name = "treatment_1.0"
    else:
        whole_df_ps = pd.concat([whole_X_df, treatment], axis=1)
        treatment_name = "treatment"

    if param["Use weighted"]:
        df_ps = compute_PS_and_IPTW(df=whole_df_ps,
                                    covariates=covariates,
                                    treatment=treatment_name)

        prob = df_ps["iptw_weight"] / df_ps["iptw_weight"].sum()
        # Resample (use original size; adjust as needed)
        n_samples = len(df_ps)
        indices = np.random.choice(len(df_ps), size=n_samples, replace=True, p=prob)

        # Create resampled arrays
        x_resampled = whole_X_df.values[indices]
        t_resampled = whole_t_df.values[indices]
        e_resampled = whole_e_df.values[indices]
        treatment_resampled = treatment.values[indices]

        whole_X_df = pd.DataFrame(x_resampled, columns=whole_X_df.columns)
        whole_t_df = pd.DataFrame(t_resampled, columns=["time"])
        whole_e_df = pd.DataFrame(e_resampled, columns=["event"])
        treatment = pd.DataFrame(treatment_resampled, columns=[treatment_name])

    df_ps = pd.concat([
        whole_X_df,
        whole_t_df,
        whole_e_df
    ],
        axis=1
    )

    if param["Include treatment"]:
        pass
    else:
        df_ps = pd.concat([df_ps, treatment], axis=1)

    dcm_wrap = DCM_Wrapper(DCM_params)

    if param["Include treatment"]:
        dcm_wrap.fit(train_set=[whole_X_df, whole_t_df.squeeze(), whole_e_df.squeeze()],
                 val_set=[whole_X_df, whole_t_df.squeeze(), whole_e_df.squeeze()])
    else:
        dcm_wrap.fit(train_set=[whole_X_df, whole_t_df.squeeze(), whole_e_df.squeeze()],
                 val_set=[whole_X_df, whole_t_df.squeeze(), whole_e_df.squeeze()])

    dcm_model = dcm_wrap.model

    train_latent = dcm_model.predict_latent_z(whole_X_df)

    train_group = np.argmax(train_latent, axis=1)

    dcm_causal_effects = plot_avg_survival_curve(df=df_ps,
                                                 group_index=train_group,
                                                 model_wrapper=dcm_wrap,
                                                 covariates=covariates,
                                                 treatment=treatment_name,
                                                 num_time=1000,
                                                 save_figure=True,
                                                 save_name=save_name,
                                                 )

    model_list.append((dcm_wrap, dcm_causal_effects))

    for i, ce in enumerate(dcm_causal_effects):
        print(f"For group {i}, the estimated causal effect: {ce}.")

    print(f"{save_name} complete")
```
